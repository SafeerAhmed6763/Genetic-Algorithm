```python
# Deficiencies
import math
import pandas
import random
import numpy as np
from __future__ import division
```

```python
Data = pandas.read_csv('DataSets/heart.csv')
size = Data.shape
sizeOfAttribute = size[0]

# Size of P etc S = 0.5 for half
S = 0.33

sizeOfP = S * sizeOfAttribute
sizeOfP = math.ceil(sizeOfP)
sizeOfP = int(sizeOfP)
Data
```

```python
# GA implementation form here
data = Data.ix[:,1].as_matrix()
data = data.flatten()
```

```python
# create population
psize = 50
rr = random.sample(range(sizeOfAttribute), sizeOfP)
pp = np.asmatrix(rr)
for f in range(psize-1):
    rr = random.sample(range(sizeOfAttribute), sizeOfP)
    r1 = np.asmatrix(rr)
    pp = np.concatenate((pp, r1), axis=0)
pt = pp
for f in range(psize):
    for t in range(sizeOfP):
        pt[f,t] = data[pt[f,t]]
ptt = pt
ptt.sort(axis=1)
```

```python
ptt
```

```python
# fitness function
ss = int (1/S)
dl = Data.shape[0]
fit = []
for k in range(psize):
    ans = 0
    t = 0
    m = 0
    for i in range(sizeOfP):
        for j in range(ss):
            ans = ans + abs(ptt[k,i] - data[t])
            if(ptt[k,i] != data[t]):
                m = m + 1
            t = t + 1
    fit.append( (1/dl) * (ans/m) )
```

```python
for f in fit :
    if (f ):
        print f
```

```python
fit  = np.asarray(fit)
bestT = np.where(fit == fit.min())
```
